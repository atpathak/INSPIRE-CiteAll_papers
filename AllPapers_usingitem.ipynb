```python
with open('INSPIRE-CiteAll.tex') as f:
    full_string = ' '.join([line.strip() for line in f])
    my_list = [string.strip() for string in full_string.split("%\\cite{")]
```

```python
my_list2 = [string.replace('''%''', '''''') for string in my_list]
my_list3 = [i[i.find('''\\bibitem{'''):] for i in my_list2]
my_list4 = [i[i.find(''' '''):] for i in my_list3]
#my_list5 = [string.replace('''}''', '''''') for string in my_list4]
my_list6 = [string.replace('''\\textit{''', '''''') for string in my_list4]
my_list7 = [string.replace('''\\textbf{''', '''''') for string in my_list6]
my_list8 = [string.replace("arXiv", "ePrint") for string in my_list7]
my_list9 = [string.replace(",", ",\n") for string in my_list8]
my_list10 = [string.replace("},\n", "") for string in my_list9]
my_list11 = [string.replace("doi:", "\n DOI:") for string in my_list10]
my_list12 = [string.replace("[ePrint:", "\n [ePrint:") for string in my_list11]
my_list13 = [string.replace("[ATLAS]", "ATLAS Collaboration") for string in my_list12]
my_list14 = [string.replace(".}", ".") for string in my_list13]
my_list15 = [string.replace('''``''', '''"''') for string in my_list14]
my_list16 = [string.replace(''',\n\'\'''', ''' Published in:''') for string in my_list15]
my_list17 = [string.replace(",\n"," ") for string in my_list16]
my_list18 = [string.replace('''Published in:''', '''" \n Published in:''') for string in my_list17]
my_list19 = [string.replace("[ePrint:", "ePrint:") for string in my_list18]
my_list20 = [string.replace("[hep-ex]].", "[hep-ex]") for string in my_list19]
my_list21 = [i[:i.find('''[hep-ex]''')] for i in my_list20]
my_list21
```

```python
for count, ele in enumerate(my_list21, 0):
    print("{}.{}".format(count, ele))
```
