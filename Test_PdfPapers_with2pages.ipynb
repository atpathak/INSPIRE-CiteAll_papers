```python
import PyPDF2
import re
```

```python
reader = PyPDF2.PdfReader("1806.00425.pdf")
num_pages = len(reader.pages)
print(num_pages)
string1 = "A. Pathak"
for p, page in enumerate(reader.pages, start=1):
    text = page.extract_text()
    res_search = re.search(string1, text)
    print(p, res_search)
```

```python
writer = PyPDF2.PdfWriter()
writer.add_page(reader.pages[0])
page2 = reader.pages[24-1]
writer.add_page(page2)
with open("test-output.pdf", "wb") as fp:
    writer.write(fp)
```

```python
string1 = "A. Pathak"
with open("test-output4.pdf", "wb") as fp:
    writer = PyPDF2.PdfWriter()
    for i in range(1, 5):
        print("reading : "+ my_list2[i] + " now")
        reader = PyPDF2.PdfReader(my_list2[i]+".pdf")
        writer.add_page(reader.pages[0])
        for p, page in enumerate(reader.pages, start=1):
            text = page.extract_text()
            res_search = re.search(string1, text)
            if (res_search):
                print("Matching on page: ", p)
                page2 = reader.pages[p-1]
                writer.add_page(page2)
    writer.write(fp)
```
